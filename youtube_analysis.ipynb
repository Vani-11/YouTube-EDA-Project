```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global YouTube Statistics.csv", encoding="latin-1")
```

```python
pd.set_option("display.float_format","{:.3f}".format)
```

```python
categories_subscribes_mean = df.groupby("category")["subscribers"].mean()
categories_subscribes_mean
```

```python
categories_subscribes_mean.sort_values(ascending=False, inplace=True)
```

```python
categories_subscribes_mean.head(7).plot(kind="bar")
plt.title("Mean Subscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.xticks(rotation=45)
plt.show()
```

```python
channel_type_counts= df["channel_type"].value_counts()
channel_type_counts
```

```python
channel_type_counts.head().plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Channel Types")
plt.show()
```

```python
country_subscribers_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending=False)
```

```python
country_subscribers_total.head(10).plot(kind="bar")
plt.title("Total Subscribers by Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxenplot(x="channel_type" ,y="highest_monthly_earnings",data=df)
plt.title("Boxen Plot of Highest Monthly Earnings by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Highest Monthly Earnings")
plt.xticks(rotation=90)
plt.show()
```

```python
correlation = df["subscribers"].corr(df["video views"])
print("Correlation:", correlation)
```

```python
df.groupby(["category", "Country"])['subscribers'].sum().unstack().plot(kind="bar", stacked=True)
plt.title("Total Subscribers by Category and Country")
plt.xlabel("Category")
plt.ylabel("Total Subscribers")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```
